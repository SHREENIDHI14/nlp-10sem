```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv',encoding='ISO-8859-1',usecols=['v1','v2'])
```

```python
df.head()
```

```python
import nltk
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize as st
from nltk.stem import WordNetLemmatizer as wordnet
import re
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
corpus=[]
wordnet=wordnet()
length = len(df['v2'])
for i in range(length):
  rev=re.sub('[^a-zA-Z]','',df['v2'][i]) #rem anything that isnt a character ie remove nos etc get only teh alphaets
  rev=rev.lower()
  rev=rev.split()
  rev =[wordnet.lemmatize(word) for word in rev if word not in stopwords.words('english')] #if stopwords will lematize other wise ignore
  rev = ' '.join(rev)
  corpus.append(rev)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500) #
x=cv.fit_transform(corpus).toarray()
y=df['v1']
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print("Accuracy",model.score(x_test, y_test))
confusion_matrix(y_test, model.predict(x_test))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```
