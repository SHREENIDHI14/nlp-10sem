```python
import pandas as pd
train=pd.read_csv('/content/drive/MyDrive/NLP-8TH SEM/Participants_Data_WH20/Participants_Data_WH20/Train.csv')
test=pd.read_csv('/content/drive/MyDrive/NLP-8TH SEM/Participants_Data_WH20/Participants_Data_WH20/Test.csv')
ss=pd.read_csv('/content/drive/MyDrive/NLP-8TH SEM/Participants_Data_WH20/Participants_Data_WH20/sample submission.csv')
```

```python
target='Labels'
```

```python
train['Text']=train['Text'] + ' ' + train['Text_Tag'].fillna("").apply(lambda x: ' '.join(x.split(',')))
test['Text']=test['Text'] + ' ' + test['Text_Tag'].fillna("").apply(lambda x: ' '.join(x.split(',')))
```

```python
from autoviml.Auto_ViML import Auto_ViML
```

```python
model, features, trainm, testm = Auto_ViML(
    train=train,
    target="Labels",
    test=test,
    sample_submission=ss,
    hyper_param="RS",
    feature_reduction=True,
    scoring_parameter="neg_log_loss",
    KMeans_Featurizer=False,
    Boosting_Flag="CatBoost",
    Binning_Flag=False,
    Add_Poly=False,
    Stacking_Flag=False,
    Imbalanced_Flag=False,
    verbose=0,
)
```

```python
test_ids = []
force_senti = []
for index,row in test.iterrows():
 temp = train[train['Text'] == row['Text']]
 #temp = temp[temp['Text_Tag'] == row['Text_Tag']]
 if temp.shape[0] > 0:
  print(temp.shape)
  force_senti.append(list(set(temp['Labels'].tolist()))[0])
  test_ids.append(index)
```
